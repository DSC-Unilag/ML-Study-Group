CLEANING THE RAW_TOTAL_FIGHT_DATA DATASET

```python
# First I import some libraries which we'll use to read and visualise our data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# i use the pandas read_csv method to read a ';' seperated file format and save it in a variable
df = pd.read_csv('raw_total_fight_data.csv', sep = ';')
# i print out the first five rows of the dataset
df.head()
```

```python
# the last five rows of the data set
df.tail()
```

```python
# The list of columns in our dataset
df.columns
```

```python
#the number of rows and columns respectively in our dataset
df.shape
```

```python
# using indexing i save the fourth row in a variable
index_4 = df[3:4]
index_4
```

```python
# statistical summary of the dataset
df.describe()
```

```python
# i dropped all rows with all empty values
df.dropna(how = 'all', inplace = True)
```

```python
#changed the date column to the appropriate format
df['date'] = pd.to_datetime(df['date'], infer_datetime_format = True)
```

```python
# heatmap showing levels of correlation amongst columns
plt.figure(figsize = (15, 8))
corr_matrix = df.corr()
sns.heatmap(data = corr_matrix, annot = True)
```
