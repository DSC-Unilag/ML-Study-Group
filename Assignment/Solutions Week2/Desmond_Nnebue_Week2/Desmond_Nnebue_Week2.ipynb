<h5>Dealing with missing values</h5>

```python
no_value = data.isnull()
no_value.head()
```

```python
for val in no_value.columns.values.tolist():
    print(f"\n{val}")
    print(f"{no_value[val].value_counts()}")
```

```python
#I have decided to remove the fighters without date of birth because we cannot assume a date of birth for them
#I also dropped the missing values in weight, reach, height and stance because they're also data that are unique to every individual fighter
data.dropna(subset=["DOB", "Height", "Reach", "Weight", "Stance"], axis=0, inplace=True)

# resetting index after dropping
data.reset_index(drop=True, inplace=True)
data.head(10)
```
